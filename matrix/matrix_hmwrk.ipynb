## Уровень 0

Создайте матрицу 3 на 3, заполненную числами от 1 до 9  
Прибавьте к ней матрицу с единицами на диагонале  
Умножьте полученную матрицу на вектор (1, 2, 3)  
Посчитайте обратную матрицу к полученной матрице

***Решение***

```python
import numpy as np

one_to_nine = np.arange(1, 10).reshape(3, 3)
diagonal_one = np.eye(3)
vector = np.array([1, 2, 3])

matrix_summ = one_to_nine + diagonal_one
multiply = matrix_summ * vector
inverse_matrix = np.linalg.inv(multiply)
```

При помощи матричных операций выполните:  
Уменьшить объект в два раза.  
Повернуть на 130 градусов.  
Отразить объект относительно прямой y=x  
Выполнить смещение объекта и масштабирование (применить 2 матрицы трансформаций). Смещение на (200, 300) масштабирование на (0.5, 1,2)

***Решение***

```python
import numpy as np
import matplotlib.pyplot as plt

A = [[-100, -100, 1], [-100, 100, 1], [100, 100, 1], [0, 0, 1], [100, -100, 1], [-100, -100, 1]]

A = np.array(A)
x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
#уменьшаем в два раза

small = A / 2
small = np.array(small)

x = small[:,0]
y = small[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
#поворот изначального объекта на 130 градусов

from math import cos, sin, pi

theta = 130 * 180/pi
TransformArray = np.array([
    [cos(theta), sin(theta), 0],
    [-sin(theta), cos(theta), 0],
    [0, 0, 1]
])

rotation = []
for point in A:
    new_point = np.dot(TransformArray,np.transpose(point))
    rotation.append(new_point)
rotation = np.array(rotation).astype(int)

plt.figure(
    figsize = (8, 8),
    )
x = rotation[:,0]
y = rotation[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
#Отразить объект относительно прямой y=x
mirror = np.array(A) * -1

x = mirror[:,0]
y = mirror[:,1]
plt.plot(x, y)
plt.ylim([-200, 500])
plt.xlim([-200, 500])
```

```python
#Смещение на (200, 300) и масштабирование на (0.5, 1,2)

#смещение
move = np.array(A) + [200, 300, 1]

x = move[:,0]
y = move[:,1]
plt.plot(x, y)
plt.ylim([0, 500])
plt.xlim([0, 500])
```

```python
# ^ и масштабирование

TransformArray = np.array([[0.5, 0, 0], [0, 1.2, 0], [0, 0, 1]])

scale = []
for point in move:
    new_point = np.dot(TransformArray,np.transpose(point))
    scale.append(new_point)
scale = np.array(scale).astype(int)

x = scale[:,0]
y = scale[:,1]
plt.plot(x, y)
plt.ylim([0, 500])
plt.xlim([0, 500])
```
